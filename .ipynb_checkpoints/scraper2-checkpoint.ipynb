```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
from splinter import Browser
from pprint import pprint
import json
import re
```

```python
index_url = "https://theinfosphere.org/Episode_Transcript_Listing"
```

```python
def soupify(url):
    browser = Browser("chrome")
    browser.visit(url)
    html_code = browser.html
    soup = bs(html_code, "html.parser")
    return soup
    browser.windows[0].close()
```

```python
def url_creation(extension):
    url = "https://theinfosphere.org" + extension
    return url
```

```python
url_list = []
```

```python
index_soup = soupify(index_url)
index_links = index_soup.select("b a")
```

```python
for i in index_links:
    extension = i["href"]
    extension = url_creation(extension)
    url_list.append(extension)
```

```python
url_list_tv = url_list[:-4]
url_list_movies = url_list[-4:]
```

```python
characters = []
quotes = []

error_list = []

loop_char_count = 0
loop_quote_count = 0

list_length = len(url_list_movies)

for i in url_list_movies:
    
    print(i)

    loop_soup = soupify(i)
    loop_lines = loop_soup.select("p")
    loop_index = url_list_movies.index(i)
    
    if loop_index == 125 \
    or loop_index == 126 \
    or loop_index == 129 \
    or loop_index == 130 \
    or loop_index == 131 \
    or loop_index == 133 \
    or loop_index == 134 \
    or loop_index == 135 \
    or loop_index == 136 \
    or loop_index == 139:
        print("We don't want this one, trust me")
        continue

    for k in loop_lines:
        k = k.text.strip()
        k = re.sub(r'\[[^)]*\]', '', k)
        k = re.sub(r'\([^)]*\)', '', k)
        k = re.sub(r'\<[^)]*\>', '', k)
        k = re.sub(r'\n', '', k)



        if len(k) > 0 and ":" in k:
            k = k.split(":")
            
            if len(k[0]) > 0 and k[0] != " " and len(k[1]) > 0:

                loop_character = k[0]
                loop_character = loop_character.strip()

                if loop_character[0] == "‚®Ç":
                    loop_character = loop_character[1:]

                characters.append(loop_character)
                loop_char_count += 1
                print("char = " + loop_character)


                loop_quote = k[1:]

                if len(loop_quote) > 1:
                    "".join(loop_quote)

                loop_quote = loop_quote[0].strip()
                quotes.append(loop_quote)
                loop_quote_count += 1
                print("quote = " + loop_quote)
                
            else:
                print(f"Something is wrong here")
        
        else:
            print(f"Empty <p> at {i}")



    if loop_char_count != loop_quote_count:
        error_list.append(url_list_movies.index(i))
        
    loop_char_list = 0
    loop_quote_list = 0
        
    print(f"Finished parsing #{loop_index} of {list_length}")
```
