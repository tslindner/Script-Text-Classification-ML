```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
from splinter import Browser
from pprint import pprint
import json
import re
```

```python
index_url = "http://www.simpsonsarchive.com/episodes.html"
```

```python
def soupify(url):
    browser = Browser("chrome")
    browser.visit(url)
    html_code = browser.html
    soup = bs(html_code, "html.parser")
    return soup
    browser.windows[0].close()
```

```python
def url_creation(extension):
    if extension.startswith("http") == False and extension.startswith("episodes/mini") == False:
        url = "http://www.simpsonsarchive.com/" + extension
        return url
```

```python
url_list = []
```

```python
index_soup = soupify(index_url)
index_links = index_soup.select("p table tbody tr td font a")
```

```python
for i in index_links:
    extension = i["href"]
    extension = url_creation(extension)
    url_list.append(extension)
```

```python
characters = []
quotes = []

list_length = len(url_list)

for i in url_list:
    
    print(i)
    
    try:
        simp_soup = soupify(i)
        test = simp_soup.text.lower()
        test2 = test.split("quotes and scene")
        test3 = test2[1].split("Contributors")
        test4 = test3[0].split("\n")
        test5 = [quote for quote in test4 if ":" in quote and quote.startswith("%") == False]

        loop_index = url_list.index(i)

        for j in test5:

            test6 = j.split(":")

            characters.append(test6[0])
            quotes.append(test6[1:])

        print(f"Finished parsing #{loop_index} of {list_length}")
        
    except:
        print(f"issue with {i}")
```

```python
# 4 characters
df_4ch = df2[(df2.Characters_stripped == "homer") | (df2.Characters_stripped == "bart") | (df2.Characters_stripped == "marge")
   | (df2.Characters_stripped == "lisa")]
```

```python
def quotes_check(x):
    result = x[0].strip()
    result = re.sub(r'\[[^)]*\]', '', result)

    return result




df_4ch['Quotes_stripped'] = df.apply(lambda row: quotes_check(row['Quotes']), axis=1)
```

```python
del df_4ch["Characters"]
del df_4ch["Quotes"]
df_4ch = df_4ch.rename(index=str, columns={"Characters_stripped": "Characters", "Quotes_stripped": "Quotes"})
```
