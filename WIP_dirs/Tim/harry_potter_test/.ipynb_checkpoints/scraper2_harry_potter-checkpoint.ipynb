```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
from splinter import Browser
from pprint import pprint
import json
import re
```

```python
index_url = "http://www.hogwartsishere.com/library/book/7391/"
```

```python
def soupify(url):
    browser = Browser("chrome")
    browser.visit(url)
    html_code = browser.html
    soup = bs(html_code, "html.parser")
    return soup
    browser.windows[0].close()
```

```python
def url_creation(extension):
    url = "http://www.hogwartsishere.com" + extension
    return url
```

```python
url_list = []
```

```python
index_soup = soupify(index_url)
index_links = index_soup.select("body div section div div div div ul li a")
```

```python
for i in index_links:
    extension = i["href"]
    extension = url_creation(extension)
    url_list.append(extension)
```

```python
characters = []
quotes = []
loop_quote = []
del_index = []

skip_bool = False
new_speaker = True
new_line = False

for i in url_list:
    loop_soup = soupify(i)
    script = str(loop_soup.get_text).split("<br/>")
    
    for j in script:

        if skip_bool == True:
            skip_bool = False
            new_line = True
            continue

        elif j.isupper() and new_speaker == True:

            skip_bool = True

            characters.append(j.strip())

        elif new_line == True:
            if j.strip() == "":
                new_line = False
                " ".join(loop_quote)
                loop_quote = " ".join(loop_quote)
                quotes.append(loop_quote)
                loop_quote = []

            else:
                loop_quote.append(j)
                print(loop_quote)
            
            
            
        
for i in range(len(quotes)):
    if type(quotes[i]) == list:
        quotes[i] = " ".join(quotes[i])
        
    quotes[i] = re.sub(r'\[[^)]*\]', '', quotes[i])

    if quotes[i].strip() == "":
        del_index.append(i)
        
for i in range(len(characters)):
    if type(characters[i]) == list:
        characters[i] = "".join(characters[i])
        
    characters[i] = re.sub(r'\[[^)]*\]', '', characters[i])



for i in del_index[::-1]:
    del quotes[i]
    del characters[i]
```
