```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup as bs
from splinter import Browser
from pprint import pprint
import json
import re
import itertools
```

```python
index_url = "http://communicationsoffice.tripod.com/index1.html"
```

```python
def soupify(url):
    browser = Browser("chrome")
    browser.visit(url)
    html_code = browser.html
    soup = bs(html_code, "html.parser")
    return soup
    browser.windows[0].close()
```

```python
def url_creation(extension):
    if extension.startswith("http") == False:
        url = "http://communicationsoffice.tripod.com/" + extension
        return url
```

```python
url_list = []
```

```python
index_soup = soupify(index_url)
index_links1 = index_soup.select("p table tbody tr td p font a")
index_links2 = index_soup.select("p table tbody tr td font p a")
```

```python
for i in index_links1:
    extension = i["href"]
    extension = url_creation(extension)
    url_list.append(extension)
    
for i in index_links2:
    extension = i["href"]
    extension = url_creation(extension)
    url_list.append(extension)
```

```python
characters = []
quotes = []
temp_characters = []
temp_quotes = []
del_index = []
error_list = []


list_length = len(url_list)

for i in url_list:
    first = True
    quote_switch = False
    inner_loop_quote = []
    loop_quote = []
    loop_characters = []
    inner_loop_index = 0
    
#     if url_list.index(i) == 0:
#         continue
    
    loop_index = url_list.index(i)
    print(i)
    
#     try:
    ww_soup = soupify(i)
    ww_soup = ww_soup.text.split("\n")        
    for j in ww_soup:
        inner_loop_index += 1

        if quote_switch == True and j.isupper() == False:
            if len(j) > 1:
                inner_loop_quote.append(j)
            else:
                quote_switch = False


        if j.isupper() and j != "THE END" and quote_switch == False:
            if j.lower().strip() == "josh" or\
            j.lower().strip() == "bartlet" or\
            j.lower().strip() == "c.j." or\
            j.lower().strip() == "toby" or\
            j.lower().strip() == "leo" or\
            j.lower().strip() == "sam" or\
            j.lower().strip() == "donna" or\
            j.lower().strip() == "abbey" or\
            j.lower().strip() == "will":
                
                main_char = j.lower().strip()
                print(main_char)

                if first == True:
                    first = False

                else:
                    loop_quote.append(inner_loop_quote)
                    print(inner_loop_quote)
                    inner_loop_quote = []
                    
                loop_characters.append(main_char)
                quote_switch = True

    loop_quote.append(inner_loop_quote)
    print(inner_loop_quote)
    inner_loop_quote = []
    
    num_of_chars = len(loop_characters)
    num_of_lines = len(loop_quote)
    
#     while len(loop_characters) > len(loop_quote):
#         del loop_characters[-1]

#     while len(loop_quote) > len(loop_characters):
#         del loop_quote[-1]

    if len(loop_characters) != len(loop_quote):
        print(f"Uh oh, we have an issue at {i}.  There are {num_of_chars} characters but {num_of_lines} lines!")
        error_list.append(i)
        continue


    temp_characters.append(loop_characters)
    temp_quotes.append(loop_quote)
    
    loop_characters = []
    loop_quote = []



    print(f"Finished parsing #{loop_index} of {list_length}")
    print(f"Ended with {num_of_lines} lines")
        
#     except:
#         print(f"issue with {i}")

        

print(len(characters), len(quotes))

quotes = list(itertools.chain.from_iterable(temp_quotes))
characters = list(itertools.chain.from_iterable(temp_characters))


# try:
for i in range(len(quotes)):
    if type(quotes[i]) == list:
        quotes[i] = " ".join(quotes[i])
        
    quotes[i] = re.sub(r'\[[^)]*\]', '', quotes[i])

    if quotes[i].strip() == "":
        del_index.append(i)
        
for i in range(len(characters)):
    if type(characters[i]) == list:
        characters[i] = "".join(characters[i])
        
    characters[i] = re.sub(r'\[[^)]*\]', '', characters[i])



for i in del_index[::-1]:
    del quotes[i]
    del characters[i]
```
