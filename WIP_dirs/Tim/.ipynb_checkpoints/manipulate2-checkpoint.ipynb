```python
def character_check(x):
    if x[:8] == "ZOIDBERG":
        return x[:8]
    
    elif x[:3] == "FRY":
        return x[:3]
    elif x[:5] == "LEELA":
        return x[:5]
    elif x[:6] == "BENDER":
        return x[:6]
    elif x[:10] == "FARNSWORTH":
        return x[:10]
    elif x[:3] == "AMY":
        return x[:3]
    elif x[:6] == "HERMES":
        return x[:6]
    elif x[:4] == "ZAPP":
        return x[:4]
    else:
        return x


df['Characters_consolidated'] = df.apply(lambda row: character_check(row['Characters']), axis=1)
```
