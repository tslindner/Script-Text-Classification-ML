```python
import pandas as pd
df = pd.read_csv("clean_output_tv.csv")
df = df.drop(["Unnamed: 0"], axis=1)
```

```python
df.groupby("Characters").count().sort_values(by="Quotes", ascending=False).head(10)
```

```python
mask = (df['Quotes'].str.len() >22)
df2 = df.loc[mask]
```

```python
def character_check(x):
    if x[:8] == "Zoidberg":
        return x[:8]
    
    elif x[:3] == "Fry":
        return x[:3]
    elif x[:5] == "Leela":
        return x[:5]
    elif x[:6] == "Bender":
        return x[:6]
    elif x[:10] == "Farnsworth":
        return x[:10]
    elif x[:3] == "Amy":
        return x[:3]
    elif x[:6] == "Hermes":
        return x[:6]
    elif x[:4] == "Zapp":
        return x[:4]
    else:
        return x


df2['Characters_consolidated'] = df.apply(lambda row: character_check(row['Characters']), axis=1)
```

```python
df2.groupby("Characters_consolidated").count().sort_values(by="Quotes", ascending=False).head(10)
```

```python
df3 = df2.groupby("Characters_consolidated").count().sort_values(by="Quotes", ascending=False).head(10)
```

```python
df2.sort_values(by=["Characters_consolidated"], ascending=False)
```

```python
# 8 characters
df_8ch = df2[(df2.Characters_consolidated == "Zoidberg") | (df2.Characters_consolidated == "Fry") 
    | (df2.Characters_consolidated == "Leela") | (df2.Characters_consolidated == "Bender")| 
    (df2.Characters_consolidated == "Farnsworth") | (df2.Characters_consolidated =='Amy') | 
    (df2.Characters == 'Hermes') | (df2.Characters == 'Zapp')]
```

```python
# 6 characters
df_6ch = df2[(df2.Characters_consolidated == "Zoidberg") | (df2.Characters_consolidated == "Fry") 
    | (df2.Characters_consolidated == "Leela") | (df2.Characters_consolidated == "Bender")| 
    (df2.Characters_consolidated == "Farnsworth") | (df2.Characters_consolidated =='Amy')]
```

```python
# 4 characters
df_4ch = df2[(df2.Characters_consolidated == "Fry") | (df2.Characters_consolidated == "Leela") | (df2.Characters_consolidated == "Bender")
   | (df2.Characters_consolidated == "Farnsworth")]
```

```python
# 2 characters
df_2ch = df2[(df2.Characters_consolidated == "Fry") | (df2.Characters_consolidated == "Bender")]
```

```python
del df_8ch["Characters"]
del df_6ch["Characters"]
del df_4ch["Characters"]
del df_2ch["Characters"]
```

```python
df_8ch = df_8ch.rename(index=str, columns={"Characters_consolidated": "Characters"})
df_6ch = df_6ch.rename(index=str, columns={"Characters_consolidated": "Characters"})
df_4ch = df_4ch.rename(index=str, columns={"Characters_consolidated": "Characters"})
df_2ch = df_4ch.rename(index=str, columns={"Characters_consolidated": "Characters"})
```

```python
df_8ch.to_csv("clean_8characters_tv.csv")
df_6ch.to_csv("clean_6characters_tv.csv")
df_4ch.to_csv("clean_4characters_tv.csv")
df_2ch.to_csv("clean_2characters_tv.csv")
```
