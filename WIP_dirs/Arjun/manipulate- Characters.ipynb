```python
import pandas as pd
df = pd.read_csv("output.csv")
df = df.drop(["Unnamed: 0"], axis=1)
```

```python
df.groupby("Characters").count().sort_values(by="Quotes", ascending=False).head(10)
```

```python
mask = (df['Quotes'].str.len() >22)
df2 = df.loc[mask]
```

```python
def character_check(x):
    if x[:8] == "ZOIDBERG":
        return x[:8]
    
    elif x[:3] == "FRY":
        return x[:3]
    elif x[:5] == "LEELA":
        return x[:5]
    elif x[:6] == "BENDER":
        return x[:6]
    elif x[:10] == "FARNSWORTH":
        return x[:10]
    elif x[:3] == "AMY":
        return x[:3]
    elif x[:6] == "HERMES":
        return x[:6]
    elif x[:4] == "ZAPP":
        return x[:4]
    else:
        return x


df2['Characters_consolidated'] = df.apply(lambda row: character_check(row['Characters']), axis=1)
```

```python
df2.groupby("Characters_consolidated").count().sort_values(by="Quotes", ascending=False).head(10)
```

```python
df3 = df2.groupby("Characters_consolidated").count().sort_values(by="Quotes", ascending=False).head(10)
```

```python
df2.sort_values(by=["Characters_consolidated"], ascending=False)
```

```python
# 8 characters
df_8ch = df2[(df2.Characters_consolidated == "ZOIDBERG") | (df2.Characters_consolidated == "FRY") 
    | (df2.Characters_consolidated == "LEELA") | (df2.Characters_consolidated == "BENDER")| 
    (df2.Characters_consolidated == "FARNSWORTH") | (df2.Characters_consolidated =='AMY') | 
    (df2.Characters == 'HERMES') | (df2.Characters == 'ZAPP')]
```

```python
# 6 characters
df_6ch = df2[(df2.Characters_consolidated == "ZOIDBERG") | (df2.Characters_consolidated == "FRY") 
    | (df2.Characters_consolidated == "LEELA") | (df2.Characters_consolidated == "BENDER")| 
    (df2.Characters_consolidated == "FARNSWORTH") | (df2.Characters_consolidated =='AMY')]
```

```python
# 4 characters
df_4ch = df2[(df2.Characters_consolidated == "FRY") | (df2.Characters_consolidated == "LEELA") | (df2.Characters_consolidated == "BENDER")
   | (df2.Characters_consolidated == "FARNSWORTH")]
```
